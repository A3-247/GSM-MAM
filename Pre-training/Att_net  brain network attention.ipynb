```python
from math import sqrt
 
import torch
import torch.nn as nn
 
class SelfAttention(nn.Module):
    def __init__(self, dim_q, dim_k, dim_v):
        super(SelfAttention, self).__init__()
        self.dim_q = dim_q
        self.dim_k = dim_k
        self.dim_v = dim_v
 
        self.linear_q = nn.Linear(dim_q, dim_k,bias=False)
        self.linear_k = nn.Linear(dim_q, dim_k,bias=False)
        self.linear_v = nn.Linear(dim_q, dim_v,bias=False)
        self._norm_fact = 1 / sqrt(dim_k)
 
    def forward(self, x):
        batch, dim_q = x.shape
        n = 1
        assert dim_q == self.dim_q
        q = self.linear_q(x) 
        k = self.linear_k(x) 
        v = self.linear_v(x)
        

        dist = torch.mm(q, k.transpose(0,1)) * self._norm_fact  # batch, n, n
        dist = torch.softmax(dist, dim=-1)  # batch, n, n
        att = torch.mm(dist, v)
        return att
```

```python
class DJPOS:
    def __init__(self):
        self.feature=0
        self.pos=0
        self.pos_ac=0

def cmp(a, b):
    if a.feature < b.feature:
        return -1
    else:
        return 1
    
def cmpA(a,b):
    if a.pos<b.pos:
        return -1
    else:
        return 1
```
